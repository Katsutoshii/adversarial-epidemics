# Epidemic simulation demo

Demonstrates the use of `EpidemicGraph` API for simulating an epidemic spreading over a network

```python
import networkx as nx
import matplotlib.pyplot as plt
from pathlib import Path
from datetime import datetime

from simulator import Epidemic, EpidemicGraph
```

### Setup epidemic parameters and network

```python
n: int = 100
k: int = 3

G: nx.Graph = nx.fast_gnp_random_graph(n=n, p=k/n)
ep: Epidemic = Epidemic(
    incubation_period=1,
    spread_fraction=1/5,
    recovery_period=2)
EG: EpidemicGraph = EpidemicGraph(
    G=G,
    ep=ep,
    infected={i: 0 for i in range(3)},
    immune=set([1, 2, 3]))

EG
```

### Run simulation

```python
while True:
    EG.show()
    if not EG.step(): break
```
